## Adding live births at 18y and 30y
I add to the dataset the number of live births when the age of the mother is 18 years old (quite young) and when it is 30 years old (with a greater possibility of having a better financial situation). 

```python
#At first I create a deep copy of the dataset created before
df1 = pd.read_csv("df1.csv")
df2 = df1.copy()
df2.head()
```

```python
#I import the dataset with the interested variables to add
data2 = pd.read_csv("live_births_ma_bo.csv")
data2.head()
```

```python
#data2_18 contains data related to the live births of women of 18 years old
data2_18 = data2[(data2.age == 'Y18') & (data2.ord_brth == 'TOTAL') & (data2.TIME_PERIOD == 2018)]
data2_18
```

```python
#I discard the unwanted columns
data2_18.drop(['DATAFLOW','LAST UPDATE','freq','unit','age','ord_brth','TIME_PERIOD','OBS_FLAG'], axis=1, inplace=True)
data2_18.rename(columns={'OBS_VALUE':'live_births_18y'}, inplace=True)
data2_18
```

```python
#I convert the column of live births to integer
data2_18 = data2_18.astype({"live_births_18y": int})
data2_18
```

```python
#I merge (left merge) the original dataframe df2 with the data2_18 above
result = pd.merge(df2, #left df
                  data2_18[['geo', 'live_births_18y']], #right df with just the columns I want
                  on='geo', #column they have in common
                  how='left'
                 )
result
```

```python
#Now I add the number of live births of women having 30 years at the moment of the birth
data2_30 = data2[(data2.age == 'Y30') & (data2.ord_brth == 'TOTAL') & (data2.TIME_PERIOD == 2018)]
data2_30.rename(columns={'OBS_VALUE':'live_births_30y'}, inplace=True)
data2_30 = data2_30.astype({"live_births_30y": int})
data2_30
```

```python
#I merge (left merge) the previous dataframe 'result' with the data2_30 above and I save it as a csv file for checkpoint
df2 = pd.merge(result, #left df
               data2_30[['geo', 'live_births_30y']], #right df with just the columns I want
               on='geo', #column they have in common
               how='left'
              )
df2
```

```python
compression_opts = dict(method='zip',
                        archive_name='df2.csv')  
df2.to_csv('df2.zip', index=False, compression=compression_opts) 
```

## Adding abortion ratio

The abortion ratio is defined as the number of abortions per 1000 live births in a given year by the Eurostat metadata. I consider for my analysis the abortion ratio in 2018. Where NaN is present, I replace it with the abortion ratio of the corresponding country of another year or with the abortion ratio taken from the website https://abort-report.eu/europe/ (site about abortion in Europe hold by Exelgyn, a global pharmaceutical company specialising in women’s healthcare). 

```python
#Deep copy of the dataset created before
df5 = pd.read_csv("df5.csv")
df6 = df5.copy()
df6.head()
```

```python
#I import the dataset with the interested variables to add
data6 = pd.read_csv("abortion_ratio.csv")
data6.head()
```
